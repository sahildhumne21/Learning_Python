###`Q1:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
result = []
for a, b in test_list:
    # check if first element already exists in result
    print(a,b)
    for i in range(len(result)):
        print(i)
        if result[i][0] == a:
            # extend that tuple with new value
            result[i] = result[i] + (b,)
            break
    else:
        # if not found, add a new tuple
        result.append((a, b))

print(result)
````

###`Q4`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
l1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]
t=0
l=0
s=0
for i in l1:
  if isinstance(i,tuple):
    t+=1
  elif isinstance(i,list):
    l+=1
  else:
    s+=1
print(f"Tuple-{t}\n List-{l}\n Set-{s} ")
````

###`Q1:` Write a program to find set of common elements in three lists using sets.
```
Input : ar1 = [1, 5, 10, 20, 40, 80]
        ar2 = [6, 7, 20, 80, 100]
        ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

Output : [80, 20]
```

````python
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]
s1=set(ar1)
s2=set(ar2)
s3=set(ar3)
a=[]
a.extend(s1.intersection(s2,s3))
print(a)
````

### `Q5`: Sort Dictionary key and values List.

**Example 1:**

Input:

```bash
{'c': [3], 'b': [12, 10], 'a': [19, 4]}
```

Output:

```bash
{'a': [4, 19], 'b': [10, 12], 'c': [3]}
```

**Example 2:**

Input:

```bash
{'c': [10, 34, 3]}
```

Output:

```bash
{'c': [3, 10, 34]}
```

````python
d = {'c': [3], 'b': [12, 10], 'a': [19, 4]}
sorted_lists_dict = {}
for key in d:
    sorted_lists_dict[key] = sorted(d[key])
final_dict = {}
for key in sorted(sorted_lists_dict.keys()):
    final_dict[key] = sorted_lists_dict[key]
print(final_dict)
````
