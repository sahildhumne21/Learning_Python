### **`Problem-1:`** Write a Python function that takes a list and returns a new list with unique elements of the first list.

**Exercise 1:**

Input:

```bash
[1,2,3,3,3,3,4,5]
```

Output:

```bash
[1, 2, 3, 4, 5]
```

````python
def unique_element(l):

  l2=[]
  for i in l:
    if i not in l2:

      l2.append(i)

  return l2
l=[1,2,3,3,3,3,4,5]

print(unique_element(l))
````

### **`Problem-2:`** Write a Python function that accepts a hyphen-separated sequence of words as parameter and returns the words in a hyphen-separated sequence after sorting them alphabetically.

**Example 1:**

Input:
```bash
green-red-yellow-black-white
```

Output:
```bash
black-green-red-white-yellow
```

````python
# Write code here
def alpha(words):
  word=words.split("-")
  word.sort()
  sorted="-".join(word)
  return(sorted)

words="green-red-yellow-black-white"
print(alpha(words))
````

### **`Problem 3:`** Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

```
Sample String : 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
Expected Output :
No. of Upper case characters :  9
No. of Lower case Characters :  47
```

````python
# Write code here
def count_case(Sample_String ):
  upper_count=0
  lower_count=0
  for i in Sample_String:
    if i.isupper():
      upper_count += 1
    elif i.islower():
      lower_count += 1
  return(upper_count,lower_count)

Sample_String ="CampusX is an Online Mentorship Program fOr EnginEering studentS."

upper,lower=count_case(Sample_String)
print("No. of Upper case characters : ",upper)
print("No. of Upper case characters :", lower)
````

### **`Problem 4:`** Write a Python program to print the even numbers from a given list.
```
Sample List : [1, 2, 3, 4, 5, 6, 7, 8, 9]
Expected Result : [2, 4, 6, 8]
```

````python
# Write code here
def find_even(num):
  result=[]
  for n in num:
    if n%2==0:
      result.append(n)
  return(result)
num=[1, 2, 3, 4, 5, 6, 7, 8, 9]
result=find_even(num)
print(result)


````

### **`Problem-6:`** Write a Python function to concatenate any no of dictionaries to create a new one.

```
Sample Dictionary :
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}
Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}
```

````python
# Write code here
def concatenate(*dicts):
  dict4={}
  for i in dicts:
    dict4.update(i)
  return(dict4)
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}


print(concatenate(dic1,dic2,dic3))
````

`Problem-7` Write a python function that accepts a string as input and returns the word with most occurence.

```
Input:
hello how are you i am fine thank you
```

```
Output
you -> 2
```

````python
# Write code here
def find(string):
  words=string.split()
  output=0
  for i in words:
    for j in string:
      if i==j:
        output += 1
  return(output)

string="hello how are you i am fine thank you"

print(find(string))
````

`Problem-8` Write a python function that receives a list of integers and prints out a histogram of bin size 10

```
Input:
[13,42,15,37,22,39,41,50]
```

```
Output:
{11-20:2,21-30:1,31-40:2,41-50:3}
```

````python
# Write code here
def histogram(input):
  dic1={}

  for i in input:
    upper=(i//10)*10+1
    lower=upper+9

    key=f'{lower}-{upper}'

    if key in dic1:
      dic1[key] += 1
    else:
      dic1[key] = 1

  return dic1

input=[13,42,15,37,22,39,41,50]
print(histogram(input))






````
